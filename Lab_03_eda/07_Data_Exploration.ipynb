# Remove low variance variables (be carefull!!!)

```python
dataset_df.describe()
```

```python
from sklearn import feature_selection

selector = feature_selection.VarianceThreshold(threshold=0.02)
selector.fit_transform(dataset_df).shape
```

- This line creates a `VarianceThreshold` object, which will be used to remove features (columns) with a variance lower than a specified threshold (in this case, `0.02`).
- Variance: In a dataset, variance is a measure of how much a feature varies across different samples. Features with very low variance don’t add much information to the model because they stay nearly constant across samples.
- The threshold of `0.02` means that any feature with a variance less than 0.02 will be removed.
- `selector.fit_transform(dataset_df)`: This line applies the `VarianceThreshold` filter to the DataFrame `dataset_df`. It will remove columns (features) that don't meet the variance threshold (i.e., features with variance below 0.02 will be eliminated).
- `.shape`: This outputs the shape of the resulting dataset after feature selection. It returns a tuple `(n_samples, n_features)` where:
    - `n_samples` is the number of rows (tumors) in the dataset.
    - `n_features` is the number of columns (features) that remain after applying the variance threshold.

```python
selector.get_support()
```

The method `selector.get_support()` in Scikit-learn is used to retrieve a boolean array indicating which features were selected after applying a feature selection method, such as `VarianceThreshold`.
- `selector.get_support()`: This method returns a boolean array where:
    - True means the feature was selected (i.e., it met the variance threshold).
    - False means the feature was not selected (i.e., it was removed due to low variance).

## Univariate distributions

We want to create a grid of histograms using Seaborn and Matplotlib to visualize how the features in the breast cancer dataset are distributed, comparing the distributions between benign (`target = 0`) and malignant (`target = 1`) tumors.

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
#seaborn and matplotlib.pyplot are used for visualization.
#%matplotlib inline ensures that plots are displayed inline in the notebook.

X=dataset_df

X0 = X[X['target']==0]   # Subset where target is 0 (benign)
X1 = X[X['target']==1]   # Subset where target is 1 (malignant)

fig, axes = plt.subplots(ncols=5, nrows=6, figsize=(20,15))
fig.tight_layout()

for i, ax in zip(range(X.columns.size), axes.flat):
    sns.histplot(X0.iloc[:,i], color="blue", element="step", ax=ax,  alpha=0.3)
    sns.histplot(X1.iloc[:,i], color="red", element="step", ax=ax,  alpha=0.3)
plt.show()
```

### Plotting the Histograms:
- For each feature in the DataFrame, the loop creates two histograms in different colors:
    - Blue for benign tumors (`X0`)
    - Red for malignant tumors (`X1`)
### Displaying the Plots: 
`plt.show()`

```python
%matplotlib inline

fig, axes = plt.subplots(ncols=5, nrows=6, figsize=(12,15))
fig.tight_layout(pad=2)

# This creates a grid of subplots with 5 columns and 6 rows, 
# yielding a total of 30 subplots (since the dataset has 30 features).
# fig.tight_layout(pad=2) ensures that the layout is 
# properly spaced to avoid overlap between subplots.


for i, col in enumerate(X.columns[:-1]):
    sns.boxplot(y = col, x = "target",data=dataset_df, orient='v', ax=axes[int(i/5),i%5])
```

- `y=col`: The feature (column) being plotted on the y-axis.
- `x="target"`: The target (benign or malignant) being used on the x-axis.
- `data=X`: The DataFrame being used for plotting.
- `orient='v'`: Specifies that the boxplot is vertical.
- `ax=axes[int(i/5), i%5]`: This places each boxplot into the correct subplot.
The indexing formula `int(i/5), i%5` helps in arranging the boxplots across the 5-column grid.

## Univariate test (BE CAREFULL)

This code performs feature selection on the breast cancer dataset using two different statistical methods: the Chi-Square test (`chi2`) and the ANOVA F-test (`f_classif`). The goal is to identify the top 5 features that are most important in classifying the target variable (`benign` vs. `malignant`).

```python
from sklearn import feature_selection
from sklearn.feature_selection import SelectKBest

X=dataset_df.iloc[:,:-1]

y=dataset.target

selector_chi=SelectKBest(feature_selection.chi2, k=5)
selector_f=SelectKBest(feature_selection.f_classif, k=5)

#Test
#classification: chi2, f_classif, mutual_info_classif
#regression: f_regression, mutual_info_regression

X_chi = pd.DataFrame(selector_chi.fit_transform(X, y),columns=X.columns[selector_chi.get_support()])
X_f = pd.DataFrame(selector_f.fit_transform(X, y),columns=X.columns[selector_f.get_support()])
```

- `X`: This line selects all the columns of the dataset except the last one (the `target` column), which contains the feature values (e.g., mean radius, mean texture).
- `y`: The `target` column contains the labels (`0` for benign, `1` for malignant).

```python
X
```

### Creating SelectKBest Objects:
- `SelectKBest`: This function selects the top `k` features (in this case, `k=5`) that are most relevant to the target variable (`y`).
    - `chi2`: The Chi-Square test is used for classification problems with categorical features or binary labels. It measures the dependency between two variables.
    - `f_classif`: This is an ANOVA F-test that is commonly used for classification problems with continuous features. It measures how well a feature can separate the classes.
- Two selectors (`selector_chi` and `selector_f`) are created using these methods.
### Fitting the SelectKBest Objects and Transforming the Dataset:
- `fit_transform(X, y)`:
    - This method first fits the selector (using fit) to the dataset to identify the top k features and then transforms the data to keep only those selected features.
- `get_support()`:
    - This method returns a boolean array indicating which features were selected (True) and which were not (False).
    - X.columns[selector_chi.get_support()]: Retrieves the names of the top 5 features selected by the Chi-Square test.
### Creating DataFrames:
- `X_chi`: Contains the top 5 features selected by the Chi-Square test.
- `X_f`: Contains the top 5 features selected by the ANOVA F-test.

```python
print(X.columns[selector_chi.get_support()])
print(X.columns[selector_f.get_support()])

# The print(X.columns[selector_chi.get_support()]) and print(X.columns[selector_f.get_support()]) statements 
# will output the names of the top 5 features selected by the Chi-Square test (chi2) and 
# the ANOVA F-test (f_classif), respectively.
```

1. `selector_chi.get_support()`:
- This returns a boolean array where each `True` corresponds to a feature that was selected by the Chi-Square test.
- `X.columns[selector_chi.get_support()]` retrieves the column names of the selected features.
2. `selector_f.get_support()`:
- Similar to the Chi-Square test, this returns a boolean array for the ANOVA F-test and retrieves the selected feature names.
