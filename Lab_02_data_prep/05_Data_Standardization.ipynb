```python
from sklearn.preprocessing import StandardScaler
'''
USING STANDARDSCALER:

Standardize features by removing the mean and scaling to unit variance

Centering and scaling happen independently on each feature by computing the relevant statistics on the samples in the training set. Mean and standard deviation are then stored to be used on later data using the transform method.

Standardization of a dataset is a common requirement for many machine learning estimators: they might behave badly if the individual features do not more or less look like standard normally distributed data (e.g. Gaussian with 0 mean and unit variance).
'''

scaler = StandardScaler()
scaler.fit(dataset_df) 
print(scaler)
```

- Standardization is a common requirement for many machine learning algorithms. Some algorithms, such as those using distance measures (e.g., K-Nearest Neighbors, SVM, or Logistic Regression), perform better when the features have a similar scale.
- Without standardization, features with larger ranges could dominate over others with smaller ranges, leading to biased models.
1. Importing `StandardScaler`:
- The code uses `StandardScaler` from the scikit-learn library to standardize the dataset's features.
- Standardization involves removing the mean and scaling each feature to have unit variance (variance of 1).
2. `scaler = StandardScaler()`:
- This initializes a `StandardScaler` object. This object will compute the mean and standard deviation for each feature in the dataset and use these values to transform the data so that each feature has a mean of 0 and a standard deviation of 1.
3. `scaler.fit(dataset_df)`:
- This line computes the mean and standard deviation for each feature in the dataset_df DataFrame.
- It doesn't yet apply the transformation; it just "learns" the statistics (mean and variance) that will be used in the next step.
4. `print(scaler)`:
- This prints information about the scaler object, which includes details about the scaling technique.
- It shows that a `StandardScaler` object has been created, ready to transform the data using the `transform()` method.

```python
scaler.mean_
```

If we run`scaler.mean_`, we'll get an array showing the mean of each feature in `dataset_df`

```python
scaler.scale_ 
```

```python
scaled_data = scaler.transform(dataset_df)
#The scaler instance can then be used on new data (e.g.TEST SET!)
```

- `scaler.transform(dataset_df)`:

    - This method applies the scaling transformation to the dataset.
    - It standardizes the features of `dataset_df` by subtracting the mean (from scaler.mean_) and dividing by the standard deviation (from `scaler.scale_`), so that each feature has a mean of 0 and a variance of 1.
    - This method uses the statistics learned from the `fit()` method (mean and standard deviation) to transform the data.

- `scaled_data`:

    - After transformation, `scaled_data` will contain the standardized values of the original dataset.
    - It is in the form of a NumPy array, where each feature now has a mean of 0 and a standard deviation of 1.

```python
scaled_df = pd.DataFrame(scaled_data)
scaled_df.columns = columns

scaled_df.head()
```

```python
'''
USING MinMaxScaler:

Transforms features by scaling each feature to a given range.

This estimator scales and translates each feature individually such that it is in the given range on the training set, i.e. between zero and one.

The transformation is given by:

X_std = (X - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0))
X_scaled = X_std * (max - min) + min
where min, max = feature_range.

This transformation is often used as an alternative to zero mean, unit variance scaling.
'''
from sklearn.preprocessing import MinMaxScaler
mm_scaler =  MinMaxScaler(copy=False,feature_range=(-1, 1))
mm_scaler.fit_transform(dataset_df)
```

```python
dataset_df.head()
```

- `MinMaxScaler` is used to scale and transform the features so that they lie within a specified range, typically between 0 and 1 by default.
- The parameter `feature_range=(-1, 1)` sets the desired range of the transformed data, meaning each feature will be scaled to fall within the range of -1 to 1 (instead of the default 0 to 1).
- `copy=False` ensures that the original data will be modified in place (not copied). This is useful for saving memory but be cautious as it will overwrite the original dataset.
- fit_transform(dataset_df): This method:
    - Fits the `MinMaxScaler` on `dataset_df`, which means it computes the minimum and maximum values for each feature in the dataset.
    - Transforms the dataset using the formula:

\begin{equation}
X_{\text {std }}=\frac{X-X_{\min }}{X_{\max }-X_{\min }}
\end{equation}

\begin{equation}
X_{\text {scaled }}=X_{\text {std }} \times(\max -\min )+\min
\end{equation}

where min and max are the bounds specified by `feature_range`. In this case, the values will be scaled to the range [-1, 1].

### Key Difference from `StandardScaler`:

- `MinMaxScaler` does not subtract the mean and divide by the standard deviation like `StandardScaler`. Instead, it scales each feature to a specific range (in this case, [-1, 1]).
- This method is useful when your data is not normally distributed and you want to transform it without changing the relative distances between feature values.
